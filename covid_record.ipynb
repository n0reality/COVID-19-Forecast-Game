```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
df = pd.read_csv('./src/csse_covid_19_daily_reports_country.csv')
df = df[::-1]
df.head()
```

```python
df.drop(['Deaths','Recovered','Active'],axis=1,inplace=True)
df.columns = ['ID','Country','Date','Confirmed']
```

```python
df[:10]
```

```python
df[70:80]
```

```python
df_US=df[df['Country']=='US/United States']
df_US['Diff'] = df_US['Confirmed'].diff()
df_US = df_US.fillna(0)
df_US.reset_index(inplace=True)
df_US.drop(['ID','index'],axis=1,inplace=True)
# df_US.to_csv('dist/USA.csv')
df_US.to_excel('dist/USA.xlsx')
df_US
```

```python
df_Global=df[df['Country']=='Total/Global']
df_Global['Diff'] = df_Global['Confirmed'].diff()
df_Global = df_Global.fillna(0)
df_Global.reset_index(inplace=True)
df_Global.drop(['ID','index'],axis=1,inplace=True)
# df_Global.to_csv('dist/Global.csv')
df_Global.to_excel('dist/Global.xlsx')
df_Global
```

```python
y_global=df_Global['Confirmed']
x_global=pd.to_datetime(df_Global['Date'])
plt.figure(figsize=(18,10))
plt.title('Covid-19 in Global')
plt.xlabel('Date')
plt.ylabel('Confirmed')
plt.plot(x_global,y_global,color='red',marker='o',label='Globol')


x_=np.linspace(0,82,83)
y_=[4*math.pow(x,3)+math.pow(x,2)+3000*x-650000 for x in x_]
plt.plot(x_global,y_,color='blue',marker='o',label='0407simulate')

plt.legend(loc='best')
plt.show()
```

```python
y_us=df_US['Confirmed']
x_us=pd.to_datetime(df_US['Date'])
plt.figure(figsize=(18,10))
plt.title('Covid-19 in US')
plt.xlabel('Date')
plt.ylabel('Confirmed')
plt.plot(x_us,y_us,color='red',marker='o',label='US')

x_=np.linspace(0,82,83)
y_=[math.pow(x,3)+50*math.pow(x,2)+2000*x-500000 for x in x_]
plt.plot(x_us,y_,color='blue',marker='o',label='0407simulate')

plt.legend(loc='best')
plt.show()
```
