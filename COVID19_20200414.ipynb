```python
import pandas as pd
import numpy as np
#from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df_covid = pd.read_excel('./src/covid_20200414.xlsx')
df_covid.columns
```

```python
df_covid = df_covid[['Name', 'Taiwan', 'Italy', 'Germany', 'USA', 'UK']]
df_covid
```

```python
true = [393,159516,131366,587337,88621]
last_week = [ 376,132547,104515,367776,51608 ]
this_week = true
```

```python
#3/24
_0324=[215,63927,30344,46378,6650]

#3/31
_0331=[322,101739,68329,164408,22141]

#4/7
_0417=[376,132547,104515,367776,51608]

#4/14
_0414=[393,159516,131366,587337,88621]
```

```python
#相對誤差MAE
#MAE
mae_loss=[]
for i in range(len(df_covid)):
    a_classmate = df_covid.loc[i]
    a_classmate_pred = [ a_classmate.Taiwan, a_classmate.Italy , a_classmate.Germany, a_classmate.USA, a_classmate.UK ]
    loss=0
    for j in range(5):
        loss+=( abs( true[j]-a_classmate_pred[j] )/true[j] )
    loss/=5
    mae_loss.append(loss)
mae_loss
```

```python
#相對誤差RMSE
#RMSE
rmse_loss=[]
for i in range(len(df_covid)):
    a_classmate = df_covid.loc[i]
    a_classmate_pred = [ a_classmate.Taiwan, a_classmate.Italy , a_classmate.Germany, a_classmate.USA, a_classmate.UK ]
    loss=0
    for j in range(5):
        loss+=( math.pow( (true[j]-a_classmate_pred[j])/true[j], 2)  )
    loss/=5
    loss=math.sqrt(loss)
    rmse_loss.append(loss)
rmse_loss
```

```python
#IAE(Increas Absolute Error)
#上週五國實際確診人數
#這週五國實際確診人數:
increase_true = abs( np.array(last_week) - np.array(this_week) )
iae_loss=[]
for i in range(len(df_covid)):
    a_classmate = df_covid.loc[i]
    a_classmate_pred = [ a_classmate.Taiwan, a_classmate.Italy , a_classmate.Germany, a_classmate.USA, a_classmate.UK ]
    loss=0
    for j in range(5):
        loss+=( abs( true[j]-a_classmate_pred[j] )/ increase_true[j] )
    loss/=5
    iae_loss.append(loss)
iae_loss
```

```python
#merge form
classmate_loss = df_covid
classmate_loss[u'MAE'] = mae_loss
classmate_loss[u'RMSE'] = rmse_loss
classmate_loss[u'IAE'] = iae_loss
classmate_loss
```

```python
#print the winner
print("相對MAE最小的人是：\t", classmate_loss.Name[classmate_loss.MAE.argmin()] )
print("相對RMSE最小的人是：\t", classmate_loss.Name[classmate_loss.RMSE.argmin()] )
print("相對IAE最小的人是：\t", classmate_loss.Name[classmate_loss.IAE.argmin()] )
```

```python
rank_tw,rank_italy,rank_germany,rank_usa,rank_uk=[],[],[],[],[]
for i in range(len(classmate_loss)):
    rank_tw.append( abs(classmate_loss.Taiwan[i]-true[0]) )
    rank_italy.append( abs(classmate_loss.Italy[i]-true[1]) )
    rank_germany.append( abs(classmate_loss.Germany[i]-true[2]) )
    rank_usa.append( abs(classmate_loss.USA[i]-true[3]) )
    rank_uk.append( abs(classmate_loss.UK[i]-true[4]) )
df_all = pd.DataFrame({'Name':classmate_loss.Name,'Taiwan':classmate_loss.Taiwan, 'Italy':classmate_loss.Italy,\
                       'Germany':classmate_loss.Germany, 'USA':classmate_loss.USA,'UK':classmate_loss.UK,\
                       'tw_rank':rank_tw,'italy_rank':rank_italy,\
                       'germany_rank':rank_germany,'usa_rank':rank_usa,'uk_rank':rank_uk})

df_all["Rank_Taiwan"] = df_all["tw_rank"].rank()
df_all["Rank_Italy"] = df_all["italy_rank"].rank()
df_all["Rank_Germany"] = df_all["germany_rank"].rank()
df_all["Rank_USA"] = df_all["usa_rank"].rank()
df_all["Rank_UK"] = df_all["uk_rank"].rank()
df_all['Rank_Sum']=df_all["Rank_Taiwan"]+df_all["Rank_Italy"]+df_all["Rank_Germany"]+df_all["Rank_USA"]+df_all["Rank_UK"]
df_all['MAE'] = classmate_loss.MAE
df_all['RMSE'] = classmate_loss.RMSE
df_all['IAE'] = classmate_loss.IAE
df_all['Rank_MAE'] = df_all['MAE'].rank()
df_all['Rank_RMSE'] = df_all['RMSE'].rank()
df_all['Rank_IAE'] = df_all['IAE'].rank()
df_all.drop(['tw_rank','italy_rank','germany_rank','usa_rank','uk_rank'],axis=1,inplace=True)
df_all
```
